```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
# Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\INPshy\Downloads\chromedriver_win32 (3)\chromedriver.exe") 
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://www.amazon.in/"
driver.get(url)

time.sleep(2)
# Asking the user to input the keywords he/she wants to search
user_inp = input('Enter the product you want to search : ')
```

```python
search_bar = driver.find_element_by_id("twotabsearchtextbox")    # Locating searc_bar by id
search_bar.clear()                                               # clearing search_bar
search_bar.send_keys(user_inp)                                   # sending user input to search bar
search_button = driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]/span/input')       # Locating search_button by xpath
search_button.click()   
```

```python
product_urls = []
start=0
end=6
for page in range(start,end):#for loop for scrapping 3 page
    url=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')       #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))                        #appending the urls in product_urls list
    nxt_button=driver.find_element_by_xpath("//li[@class='a-selected']/a")     #scraping the list of buttons from the page
    nxt_button.get_attribute('href')
    time.sleep(2)
images = driver.find_elements_by_xpath('//img[@class="image item itemNo0 maintain-height selected"]')
img_urls = []
img_data = []
for image in images:
    source= image.get_attribute('src')
    if source is not None:
        if(source[0:5] == 'https'):
            img_urls.append(source)
            
            
for i in range(len(img_urls)):
    if i >= 200:
        break
    print("Downloading {0} of {1} images" .format(i, 200))
    response= requests.get(img_urls[i])
    file = open(r"C:\Users\INPshy\Desktop\DATA Science\image classification"+str(i)+".jpg", "wb")
    file.write(response.content)
```

```python
# Collecting url's
urls = []

base_url = "https://www.amazon.in/s?k=Sarees&page={}&qid=1632112444&ref=sr_pg_{}"

for i in range(2,10):
    
    driver.get(base_url.format(i,i))
    
    for url in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
        
        urls.append(url.get_attribute("href"))
```

```python
# Grabbing image source

images_sarees = []

for link in urls:
    
    driver.get(link)
    time.sleep(1)
    
    try:
        
        img = driver.find_element_by_css_selector("#landingImage")
    
        source= img.get_attribute('src')
    
        images_sarees.append(source)
        
    except NoSuchElementException:
        
        images_sarees.append("Image Not Found!")
```

```python
import requests 
for i in range(len(images_sarees)):
    
    print("Downloading....")
    
    response= requests.get(images_sarees[i])
    
    file = open(r"C:\Users\INPshy\Desktop\DATA Science\Saree\Saree"+str(i)+".jpg", "wb")
    
    file.write(response.content)    
```
